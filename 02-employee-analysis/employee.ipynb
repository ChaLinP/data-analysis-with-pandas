# Handle Missing Values

````python
employee['Age'] = employee['Age'].fillna(employee['Age'].mean())
````

````python
employee['Experience']= employee['Experience'].fillna(employee['Experience'].mean())
````

# Department Salary Report

````python
report = employee.groupby("Department").agg({
    "Salary":["mean","max","min"],
    "Employee ID":"count"
})
````
