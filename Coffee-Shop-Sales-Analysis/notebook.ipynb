# Weekly Coffee Sales Analysis

This project analyzes weekly coffee sales data using Pandas.

The objectives are to:

- Explore the dataset
- Calculate total sales
- Analyze sales by coffee type
- Analyze daily sales performance
- Calculate revenue
- Compare weekday and weekend sales
- Measure sales target achievement

## Import Required Libraries

Import the libraries required for data analysis.

```python
import pandas as pd
```

## Load Dataset

Read the coffee sales dataset into a Pandas DataFrame.

```python
coffee = pd.read_csv("coffee.csv")
```

```python
coffee
```

## Dataset Overview

Display the dataset and inspect its structure.

```python
coffee.shape
```

## Total Sales Analysis

Calculate the total and average number of coffee units sold during the week.

```python
# Total Units Sold
coffee["Units Sold"].sum()
```

```python
# Average Units Sold
coffee["Units Sold"].mean()
```

## Coffee Type Analysis

Analyze sales performance by coffee type to determine which beverage generated the most sales.

```python
coffee.groupby("Coffee Type")["Units Sold"].sum()
```

```python
coffee.groupby("Coffee Type").agg({'Units Sold': 'sum'})
```

## Daily Sales Analysis

Calculate total units sold per day and identify the best-performing sales day.

```python
coffee.groupby("Day").agg({'Units Sold':'sum'})
```

```python
coffee.groupby("Day")["Units Sold"].sum()
```

## Best Sales Day

Sort daily sales in descending order to determine the highest-performing day.

```python
coffee.groupby("Day")["Units Sold"].sum().sort_values(ascending=False)
```

```python
coffee.groupby("Day").agg({'Units Sold':'sum'}).sort_values(
    by="Units Sold",
    ascending=False
)
```

```python
coffee.groupby(["Coffee Type"]).agg({'Units Sold':'sum'}).sort_values(
    by="Units Sold",
    ascending=False
)
```

## High Volume Sales Records

Filter records where more than 30 units were sold.

```python
coffee[coffee["Units Sold"] > 30]
```

## Revenue Calculation

Assign prices to each coffee type and calculate revenue for every transaction.

```python
import numpy as np
coffee['Price'] = np.where(coffee['Coffee Type'] == 'Espresso',500,700)
```

```python
coffee['Revenue'] = coffee['Units Sold'] * coffee['Price']
```

```python
coffee
```

## Revenue Analysis

Analyze total revenue and revenue distribution across coffee types and days.

```python
coffee['Revenue'].sum()
```

```python
coffee.groupby('Coffee Type').agg({'Revenue':'sum'})
```

```python
coffee.groupby('Day').agg({'Revenue':'sum'})
```

## Weekend vs Weekday Analysis

Classify records as weekday or weekend sales and compare total units sold.

```python
weekend = ((coffee['Day']=='Saturday') | (coffee['Day']=='Sunday') )

coffee["Type"] = np.where(weekend,'Weekend','Weekday')
```

```python
coffee
```

```python
coffee.groupby("Type")["Units Sold"].sum()
```

## Sales Target Achievement

Determine whether each sales record achieved the target of more than 30 units sold.

```python
coffee['Target Met'] = np.where(coffee['Units Sold'] > 30 , 'Yes', 'No')
```

```python
coffee
```

## Revenue Contribution Analysis

Calculate the percentage contribution of each coffee type to total revenue.

```python
v = coffee['Revenue'].sum()

coffee.groupby('Coffee Type')['Revenue'].sum() / v * 100
```

## Export Summary Report

Generate and export a summary report for future analysis.

```python
summary = coffee.groupby("Coffee Type").agg({
    "Units Sold": "sum",
    "Revenue": "sum"
})

summary.to_csv("weekly_coffee_report.csv")
```
