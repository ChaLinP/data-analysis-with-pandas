# Traditional GroupBy

- Find revenue by region:

```python
sales.groupby('Region')['Revenue'].sum()
```

- Find revenue by salesperson:

```python
sales.groupby('Salesperson')['Revenue'].sum()
```

# Pivot Table

```python
pd.pivot_table(sales, index='Region', values='Revenue', aggfunc='sum')
```

```python
region_revenue = pd.pivot_table(sales, index='Region', values='Revenue', aggfunc='sum')
```

```python
region_revenue.sort_values(by='Revenue',ascending=False)
```

Which region generated the highest revenue? West

```python
pd.pivot_table(sales, index='Salesperson', values='Revenue', aggfunc='sum')
```

# Region vs Category

```python
pd.pivot_table(sales, index='Region', columns='Category', values='Revenue', aggfunc='sum')
```

# Salesperson Dashboard

```python
pd.pivot_table(sales,index='Salesperson',values=['Revenue','Units Sold'],aggfunc='sum')
```

- Who generated the highest revenue? John
- Who sold the most units? John and Mike

# Product Dashboard

```python
pd.pivot_table(sales,index='Product',values=['Units Sold', 'Revenue'], aggfunc='sum')
```

- Best product by revenue? laptop
- Best product by units sold? Mouse

# Multiple Dimensions

```python
pd.pivot_table(sales,index='Region',columns='Product',values='Revenue',aggfunc='sum',fill_value=0)
```

# Multiple Aggregations

```python
pd.pivot_table(sales,index='Region',columns='Category',values='Revenue',aggfunc=['sum','mean','max'],fill_value=0)
```

# Crosstab
A crosstab counts occurrences.

```python
pd.crosstab(sales['Region'],sales['Category'])
```

- Which region sold more Electronics products? North,South and West
- Which region had more Accessories transactions? East

# Advanced Crosstab

```python
pd.crosstab(sales['Salesperson'],sales['Category'])
```

# Add Margins

```python
pd.pivot_table(sales,index='Region',columns='Category',values='Revenue',aggfunc='sum',fill_value=0,margins=True)
```
