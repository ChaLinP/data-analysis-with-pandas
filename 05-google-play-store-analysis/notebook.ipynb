# Additional Questions

1. Do higher-rated apps have more positive reviews?
2. Which category has the most positive sentiment?

For each app, what is the average rating associated with each sentiment?


```python
pd.pivot_table(playstore_reviews,index='App',values='Rating',columns='Sentiment',aggfunc='mean')
```

### Output

| App   | Negative | Positive |
|--------|----------|----------|
| App A | 4.2 | 4.2 |
| App B | 3.8 | 3.8 |
| App C | 4.6 | 4.6 |

### Notice something interesting

For a given app, the rating is usually the same for every review of that app.

If **App A** has a rating of **4.2**, then:

- Positive reviews → **4.2**
- Negative reviews → **4.2**
- Neutral reviews → **4.2**

This happens because the **rating belongs to the app**, not to the individual review.

Therefore, this pivot table often doesn't provide much additional insight, since the average rating remains the same across different sentiment categories for the same app.

**Summary**
- Rating is given by the app store (overall score)
- Sentiment is derived from user review text

So that doesn't give you the answer to the "Do higher-rated apps have more positive reviews?" question

```python
pd.crosstab(playstore_reviews['App'],playstore_reviews['Sentiment']).sort_values(by='Positive',ascending=False)
```
