# Handle Missing

The maximum rating is 19.0, which is outside the valid rating range (1–5). This indicates a data quality issue and the row should be examined before calculating statistics such as the mean or filling missing values.

```python
playstore[playstore['Rating'] > 5]
```

```python
playstore.iloc[1]
```

**Data Quality Issue**

An invalid record was identified where the columns were shifted, causing incorrect values such as:

- Category = 1.9
- Rating = 19.0

Since Google Play Store ratings must be between 1 and 5 and categories should contain text values, this row was determined to be corrupted and was removed from the dataset before further analysis.

```python
playstore = playstore.drop(10472)
```

```python
playstore.describe()
```

```python
playstore.isna().sum()
```

```python
playstore['Type'].value_counts()
```

```python
playstore[playstore['Type'].isna() == True]
```

```python
playstore = playstore.dropna(subset=['Type', 'Current Ver', 'Android Ver'])
```

```python
playstore.isna().sum()
```

```python
playstore.shape
```

```python
playstore['Rating'] = playstore['Rating'].fillna(playstore['Rating'].mean())
```
